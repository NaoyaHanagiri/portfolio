```python
# 必要部分のみ切り取り
riyoufuan_todou_2=riyoufuan.iloc[91:139,1:4]
```

```python
# 都道府県毎のSNS利用率/SNS利用に対する不安率
x = riyou_todoufuken_2["SNS利用率"]
y = riyoufuan_todou_2["不安を感じる"]
 
# 目盛の値
x_scale = 5
y_scale = 5
xscale = np.arange(40, (max(x)+1)*x_scale, x_scale)
yscale = np.arange(40, (max(y)+1)*y_scale, y_scale)

fig = plt.figure(dpi=100, figsize=(4.6,3.8))
fig.suptitle('各都道府県毎', fontsize=15, x=0.27, y=0.87)
ax = fig.add_subplot(111)
ax.yaxis.set_label_coords(-0.07,0.4)
 
# 散布図を描画
plt.scatter(x, y, c='#00FFFF', edgecolors="#37BAF2")
plt.title("都道府県別SNS利用率/SNS利用不安率", fontsize=14, fontweight="bold")
plt.xticks(xscale, fontsize=8)
plt.yticks(yscale, fontsize=8)
plt.xlabel("SNS利用率[%]", fontsize=10)
plt.ylabel("不\n安\nを\n感\nじ\nる\n[%]", fontsize=10, rotation=0)
plt.xlim(50, 85)
plt.ylim(50.0, 85)

# 回帰分析　線形
mod = LinearRegression()
df_x = pd.DataFrame(x)
df_y = pd.DataFrame(y)

# 線形回帰モデル、予測値、R^2を評価
mod_lin = mod.fit(df_x, df_y)
y_lin_fit = mod_lin.predict(df_x)
r2_lin = mod.score(df_x, df_y)
ax.text(78.0, 70, '$ R^{2} $=' + str(round(r2_lin, 4)))
plt.plot(df_x, y_lin_fit, color = '#000000', linewidth=0.5)
         
plt.gca().spines['right'].set_visible(False)
 
plt.subplots_adjust(left=0.1, right=0.98, bottom=0.13, top=0.89)
plt.savefig("【散布図】都道府県毎利用率_不安率.png", facecolor="white")
```

```python
num_cluster = range(1,10) # x
sum_of_squares = [] # yの箱
change =[] # クラスタが1つの時のsum of squaresを100としたときの比率の箱

# クラスタ数が1の時のSum of squaresを計算しておく
km1 = KMeans(n_clusters=1,random_state=0).fit(df_array)
first = km1.inertia_

# for文で想定クラスタ数分計算して箱に入れていく
for i  in num_cluster:
    km = KMeans(n_clusters=i, random_state=0).fit(df_array)
    sum_of_squares.append(km.inertia_)
    change.append(km.inertia_/first*100)

# Dataframeに結果を格納
df_5 = pd.DataFrame({"sum_of_squares":sum_of_squares,
                   "change":change}, index=num_cluster)

#描画
plt.title("エルボー法によるクラスタ数決定", fontsize=14, fontweight="bold")
plt.plot(num_cluster,sum_of_squares,marker='o')
plt.xlabel('クラスタ数')
plt.ylabel("クラスタセンタと各点の二乗総和")
plt.savefig("クラスタ数決め（エルボー法）.png", facecolor="white")
```
